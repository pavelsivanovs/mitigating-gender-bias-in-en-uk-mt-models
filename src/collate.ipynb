```python
import pandas as pd

sources = pd.read_csv("data/sources.tsv", index_col="id", sep="\t")
sources.head()
```

```python
translations = pd.DataFrame(
    columns=['template_type', 'occupation', 'source_gender', 'source', 'model', 'pipeline', 'target', 'target_gender'])
```

Zero-shot translations


```python
mbart_zeroshot = pd.read_csv("../data/mbart.zeroshot.translations.tsv", index_col="id", sep="\t")
mbart_zeroshot['model'] = 'mbart'
mbart_zeroshot['pipeline'] = 'zeroshot'
mbart_zeroshot['target'] = mbart_zeroshot['mbart_hypothesis']
mbart_zeroshot['target_gender'] = None
mbart_zeroshot.drop(columns=['mbart_hypothesis'], inplace=True)
translations = pd.concat([translations, mbart_zeroshot], ignore_index=True)
translations.head()
```

```python
opusmt_zeroshot = pd.read_csv("../data/opusmt.zeroshot.translations.tsv", index_col="id", sep="\t")
opusmt_zeroshot['model'] = 'opusmt'
opusmt_zeroshot['pipeline'] = 'zeroshot'
opusmt_zeroshot['target'] = opusmt_zeroshot['opusmt_hypothesis']
opusmt_zeroshot['target_gender'] = None
opusmt_zeroshot.drop(columns=['opusmt_hypothesis'], inplace=True)
translations = pd.concat([translations, opusmt_zeroshot], ignore_index=True)
translations.head()
```

Gender tags

```python
mbart_tagged = pd.read_csv('../data/mbart.tagged.translations.tsv', index_col=None, sep="\t")
tagged = mbart_zeroshot.copy()
tagged['target'] = mbart_tagged['mbart_hypothesis'].values
tagged['pipeline'] = 'tagged'
translations = pd.concat([translations, tagged], ignore_index=True)
translations.head(10)
```

```python
opusmt_tagged = pd.read_csv('../data/opusmt.tagged.translations.tsv', index_col=None, sep="\t")
tagged = opusmt_zeroshot.copy()
tagged['target'] = opusmt_tagged['opusmt_hypothesis'].values
tagged['pipeline'] = 'tagged'
translations = pd.concat([translations, tagged], ignore_index=True)
translations.head(10)
```

LLM corrected


```python
mbart_llm = pd.read_csv("../data/mbart.llm_corrected.translations.tsv", sep="\t")
mbart_llm['model'] = 'mbart'
mbart_llm['pipeline'] = 'llm'
mbart_llm['target'] = mbart_llm['mbart_llm_corrected']
mbart_llm['target_gender'] = None
mbart_llm.drop(columns=['mbart_hypothesis', 'mbart_llm_corrected'], inplace=True)
translations = pd.concat([translations, mbart_llm], ignore_index=True)
translations.head()
```

```python
opusmt_llm = pd.read_csv("../data/opusmt.llm_corrected.translations.tsv", sep="\t")
opusmt_llm['model'] = 'opusmt'
opusmt_llm['pipeline'] = 'llm'
opusmt_llm['target'] = opusmt_llm['opusmt_llm_corrected']
opusmt_llm['target_gender'] = None
opusmt_llm.drop(columns=['opusmt_hypothesis', 'opusmt_llm_corrected'], inplace=True)
translations = pd.concat([translations, opusmt_llm], ignore_index=True)
translations.head()
```

```python
translations.to_csv('translations.tsv', sep='\t', index=False)
```
