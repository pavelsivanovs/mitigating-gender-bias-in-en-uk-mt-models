```python
female_words = [
    # Pronouns
    "she", "her", "hers", "herself",

    # General gendered nouns
    "woman", "women", "lady", "ladies", "female", "girl", "girls",
    "mother", "mom", "mommy", "mum", "mama", "grandmother", "grandma", "granny",
    "daughter", "niece", "sister", "aunt", "queen", "princess", "duchess",

    # Gendered roles / professions
    "actress", "waitress", "stewardess", "goddess", "heroine", "policewoman",
    "chairwoman", "businesswoman", "saleswoman", "spokeswoman", "craftswoman",
    "congresswoman", "sportswoman", "countrywoman", "housewife", "midwife",
    "nun", "ballerina", "maid", "nanny", "seamstress", "bridesmaid",

    # Relationship terms
    "wife", "girlfriend", "fiancée", "bride", "mistress", "widow",
    "bachelorette", "stepmother", "stepdaughter", "mother-in-law",
    "sister-in-law", "auntie",

    # Common female first names (sample)
    "anna", "mary", "maria", "elizabeth", "sophia", "emma", "isabella",
    "olivia", "ava", "mia", "amelia", "grace", "claire", "lucy", "ella", "anne",

    # Adjectives and associations
    "feminine", "maternal", "nurturing", "gentle", "graceful", "pretty",
    "beautiful", "elegant", "delicate", "soft", "caring", "emotional",
    "compassionate", "empathetic", "supportive", "tender", "modest", "kind",

    # Cultural or contextual markers
    "makeup", "lipstick", "skirt", "dress", "heels", "handbag", "purse",
    "earrings", "necklace", "bracelet", "nail polish", "blouse", "feminism",
    "period", "pregnant", "menstruation", "ovulation", "breast", "uterus",
    "ovary", "midwife", "hairdresser", "nurse", "secretary", "teacher",
    "matron", "bride-to-be", "queen mother", "female leader", "womanhood",

    # Other relational and abstract forms
    "girlhood", "sisterhood", "motherhood", "feminist", "goddesslike", "ladylike",
    "heroines", "damsel", "diva", "belle"
]

male_words = [
    # Pronouns
    "he", "him", "his", "himself",

    # General gendered nouns
    "man", "men", "male", "boy", "boys", "gentleman", "gentlemen",
    "father", "dad", "daddy", "grandfather", "grandpa", "son", "nephew",
    "brother", "uncle", "king", "prince", "duke",

    # Gendered roles / professions
    "actor", "waiter", "steward", "god", "hero", "policeman",
    "chairman", "businessman", "salesman", "spokesman", "craftsman",
    "congressman", "sportsman", "countryman", "fireman", "mailman",
    "fisherman", "repairman", "workman", "watchman", "deliveryman",
    "hunter", "soldier", "monk", "priest", "carpenter", "blacksmith",

    # Relationship terms
    "husband", "boyfriend", "fiancé", "groom", "widower", "bachelor",
    "stepfather", "stepson", "father-in-law", "brother-in-law", "uncle",

    # Common male first names (sample)
    "john", "michael", "james", "robert", "david", "william", "alexander",
    "daniel", "matthew", "joseph", "george", "thomas", "charles", "henry",
    "andrew", "peter", "paul", "mark", "lucas", "benjamin", "tom",

    # Adjectives and associations
    "masculine", "paternal", "strong", "brave", "handsome", "tough", "rugged",
    "stoic", "confident", "dominant", "protective", "assertive", "ambitious",
    "independent", "rational", "powerful", "manly", "muscular", "heroic",

    # Cultural or contextual markers
    "suit", "tie", "beard", "mustache", "football", "beer", "barber",
    "garage", "toolbox", "hammer", "engineer", "mechanic", "builder",
    "soldier", "warrior", "captain", "pilot", "cowboy", "hunter", "boss",
    "gentleman", "leader", "patriarch", "kingdom", "bro code",

    # Other relational and abstract forms
    "manhood", "brotherhood", "fatherhood", "masculinity", "male leader",
    "heroism", "macho", "dude", "guy", "bloke", "chap"
]
```

```python
def detect_gender(text):
    t = str(text).lower()
    words = [w.strip(".,!?;:") for w in t.split()]
    if any(w in words for w in female_words):
        return "female"
    elif any(w in words for w in male_words):
        return "male"
    else:
        return "neutral"
```

```python
import pandas as pd

df = pd.read_csv('../data/translations.tsv', sep='\t')

df["Tag"] = df["source"].apply(detect_gender)
df["source"] = "<" + df["Tag"] + "> " + df["source"]

df[["source", ]].to_csv("test_tagged.csv", index=False)
print(df["Tag"].value_counts())
```
