```python
!pip install sacrebleu

import csv
import sacrebleu

input_files = [
    "/content/drive/MyDrive/machine_translation/opus_gold_tagged.tsv",
    "/content/drive/MyDrive/machine_translation/opus_gold_llm.tsv",
    "/content/drive/MyDrive/machine_translation/opus_gold_zeroshot.tsv",
    "/content/drive/MyDrive/machine_translation/mbart_gold_tagged.tsv",
    "/content/drive/MyDrive/machine_translation/mbart_gold_llm.tsv",
    "/content/drive/MyDrive/machine_translation/mbart_gold_zeroshot.tsv"
]

for input_file in input_files:
    refs = []
    hyps = []

    with open(input_file, encoding="utf-8") as f:
        reader = csv.DictReader(f, delimiter="\t")
        for row in reader:
            refs.append(row["ref"])
            hyps.append(row["mt"])

    bleu = sacrebleu.corpus_bleu(hyps, [refs])

    output_file = input_file.replace(".tsv", "_bleu.txt")
    with open(output_file, "w") as f:
        f.write(str(bleu) + "\n")

    print(f"{input_file}: {bleu.score:.2f}")
```
