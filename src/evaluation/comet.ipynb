GOLD v OPUS tagged

```python
model_path = download_model("Unbabel/wmt22-cometkiwi-da")
model = load_from_checkpoint(model_path)

input_file = "/content/drive/MyDrive/machine_translation/opus_gold_tagged.tsv"
data = []
with open(input_file, encoding="utf-8") as f:
    reader = csv.DictReader(f, delimiter="\t")
    for row in reader:
        data.append({
            "src": row["src"],
            "mt": row["mt"],
            "ref": row["ref"]
        })

model_output = model.predict(data, batch_size=8, gpus=1)
print (model_output)

output_file = input_file.replace(".tsv", "_scores.txt")
with open(output_file, "w") as f:
    for score in model_output.scores:
        f.write(str(score) + "\n")
    f.write(f"\nSystem score: {model_output.system_score}\n")
```
