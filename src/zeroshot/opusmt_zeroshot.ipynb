```python
from transformers import pipeline

translator = pipeline("translation_en_to_uk", model="Helsinki-NLP/opus-mt-en-uk")
```

```python
text1 = "We all knew we were risking our lives — the teacher , the students and our parents ."
translated = translator(text1)[0]['translation_text']
print("EN:", text1)
print("UK:", translated)
```

```python
import pandas as pd

df = pd.read_csv('../data/translations.tsv', sep='\t', index_col=None)

display(df.columns)
```

```python
import tqdm

english_source = df['source']
zero_shot_translations = []

# loop through all the english reference texts from our grepped files and put into zero-shot translation pipeline.
# put into a list form so that it could be added easily to overall df.
for english_text in tqdm.tqdm(english_source):
    translated = translator(english_text)[0]['translation_text']
    zero_shot_translations.append(translated)

zero_shot_translations
```

```python
# put df into excel file

df['opusmt_hypothesis'] = zero_shot_translations
df.to_csv('../data/opusmt.zeroshot.translations.tsv', sep='\t', index=False)
```
